```python
# Assuming that the magnetic field correlation length 
# is far greater than the electrons displacement
# + effect happen close to the source z source >> z source - z effect.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
E = np.logspace(-1.25, 0.75, 10) # TeV
tau = 10 # is 10 right?
z_mrk501 = 0.033
z_mrk421 = 0.03
B0 = np.array([5, 10, 20, 40])*10**(-15) # G
```

```python
E = E[:, np.newaxis] # column
B0 = B0[np.newaxis, :] # row

print(E.shape)
B0.shape
```

```python
theta_mrk421 = 0.5*(1+z_mrk421)**(-2)*(tau/10)**(-1)*(E/0.1)**(-1)@(B0/10**(-14))
theta_mrk501 = 0.5*(1+z_mrk501)**(-2)*(tau/10)**(-1)*(E/0.1)**(-1)@(B0/10**(-14))
```

```python
theta_mrk421.shape
```

```python
for i, b in enumerate(B0[0,:]):
    plt.plot(np.log10(E), theta_mrk501[:, i], label=f"B0 = {b} G")

plt.xlabel("log(E/TeV)")
plt.ylabel("θ, deg")
plt.ylim([0.1, 0.5])
plt.legend()  # Add legend separately
plt.grid(True, alpha=0.3)
plt.show()
```

```python
for i, b in enumerate(B0[0,:]):
    plt.plot(np.log10(E), theta_mrk421[:, i], label=f"B0 = {b} G")

plt.xlabel("log(E/TeV)")
plt.ylabel("θ, deg")
plt.ylim([0.1, 0.5])
plt.legend()  # Add legend separately
plt.grid(True, alpha=0.3)
plt.show()
```
